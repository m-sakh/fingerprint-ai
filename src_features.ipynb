```python
"""
features.py

This file converts raw writing into numerical features.
These features are the first version of the author's "fingerprint."
"""

import re
import numpy as np
```

```python
def extract_features(text: str) -> np.ndarray:
    """
    Extract simple stylometric features from one writing sample.

    Input:
        text: raw writing sample as a string

    Output:
        numpy array of numerical features
    """

    # Split the text into rough sentence units using punctuation.
    sentences = re.split(r"[.!?]+", text)

    # Extract words and convert them to lowercase for consistency.
    words = re.findall(r"\b\w+\b", text.lower())

    # Count how many words appear in each sentence.
    sentence_lengths = [
        len(re.findall(r"\b\w+\b", sentence))
        for sentence in sentences
        if sentence.strip()
    ]

    # Count the length of each word.
    word_lengths = [len(word) for word in words]

    # Avoid division by zero for very short or empty files.
    word_count = max(len(words), 1)
    sentence_count = max(len(sentence_lengths), 1)

    # Return a numerical fingerprint of the text.
    return np.array([
        len(words),                                           # Total word count
        np.mean(sentence_lengths) if sentence_lengths else 0, # Average sentence length
        np.std(sentence_lengths) if sentence_lengths else 0,  # Sentence length variation
        np.mean(word_lengths) if word_lengths else 0,         # Average word length
        text.count(",") / word_count,                        # Comma usage rate
        text.count(";") / word_count,                        # Semicolon usage rate
        text.count(":") / word_count,                        # Colon usage rate
        text.count("—") / word_count,                        # Em dash usage rate
        text.count("?") / sentence_count,                    # Question usage rate
        len(set(words)) / word_count,                         # Vocabulary diversity
    ])
```
